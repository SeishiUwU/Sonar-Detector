# **Importing Dependencies:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# **Data Collection & Preprocessing:**

```python
#Loading Data
sonar = pd.read_csv("/kaggle/input/sonar-data/sonardata.csv", header = None)
#Displaying the first 5 lines of the dataset
sonar.head()
```

```python
#Number of rows & columns
sonar.shape
```

```python
#Showing statistical measures of the data
sonar.describe()
```

```python
#Counting how many rocks and mines are there
sonar[60].value_counts()
```

M => Mine

R => Rock

```python
#Seperating data and labels
X = sonar.drop(columns = 60, axis = 1)
Y = sonar[60]
```

```python
X
```

```python
Y
```

# Training and Test Data:

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

# Model Training (Logistic Regression):

```python
model = LogisticRegression()
```

# Training the model with the training data

```python
model.fit(X_train, Y_train)
```

# Model Evaluation:

# Accuracy:

```python
X_train_pred = model.predict(X_train)
train_acc = accuracy_score(X_train_pred, Y_train)
```

```python
train_acc
```

```python
X_test_pred = model.predict(X_test)
test_acc = accuracy_score(X_test_pred, Y_test)
```

```python
test_acc
```
